```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
T=float(input("Enter the Total Time you wish to monitor:"))
q1=eval(input("Enter the initial(t=0) value of angle q1 in radians:"))
q2=eval(input("Enter the initial(t=0) value of angle q2 in radians:"))
q1f=eval(input("Enter the final(t=T) value of angle q1 in radians:"))
q2f=eval(input("Enter the final(t=T) value of angle q2 in radians:"))
```

```python
t=np.linspace(0,T,50*int(T))
q1t=q1+(q1f-q1)*(t/T)
q2t=q1+(q2f-q2)*(t/T)
plt.plot(t,q1t,label="q1(t)")
plt.plot(t,q2t,label="q2(t)")
plt.xlabel("Time")
plt.ylabel("Joint Angle")
plt.title("Linear Joint-Space Trajectory")
plt.legend()
plt.grid(True)
plt.show()
```

```python
t=np.linspace(0,T,50*int(T))
a0=q1
a1=0
a2=(3*(q1f-q1))/(T**2)
a3=(2*(q1-q1f))/(T**3)
a_0=q2
a_1=0
a_2=(3*(q2f-q2))/(T**2)
a_3=(2*(q2-q2f))/(T**3)

q1t=a0+a1*t+a2*t**2+a3*t**3
q1vt=a1+2*a2*t+3*a3*t**2
q1at=2*a2+6*a3*t
q1jt=6*a3*np.ones_like(t)

q2t=a_0+a_1*t+a_2*t**2+a_3*t**3
q2vt=a_1+2*a_2*t+3*a_3*t**2
q2at=2*a_2+6*a_3*t
q2jt=6*a_3*np.ones_like(t)

plt.plot(t,q1t,label="q1(t)")
plt.plot(t,q2t,label="q2(t)")
plt.xlabel("Time")
plt.ylabel("Joint Angle")
plt.title("Cubic Joint-Space Trajectory")
plt.legend()
plt.grid(True)
plt.show()
plt.plot(t,q1vt,label="v1(t)")
plt.plot(t,q2vt,label="v2(t)")
plt.xlabel("Time")
plt.ylabel("Joint Velocities")
plt.title("Joint-Space Velocity Graphs")
plt.legend()
plt.grid(True)
plt.show()
plt.plot(t,q1at,label="a1(t)")
plt.plot(t,q2at,label="a2(t)")
plt.xlabel("Time")
plt.ylabel("Joint Accelerations")
plt.title("Joint-Space Acceleration Graphs")
plt.legend()
plt.grid(True)
plt.show()
plt.plot(t,q1jt,label="jerk1(t)")
plt.plot(t,q2jt,label="jerk2(t)")
plt.xlabel("Time")
plt.ylabel("Joint Jerks")
plt.title("Joint-Space Jerk Graphs")
plt.legend()
plt.grid(True)
plt.show()
```
