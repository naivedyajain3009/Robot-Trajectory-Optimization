```python
import numpy as np
import matplotlib.pyplot as plt

def computation_plot(q1,q2):
  l1=1
  l2=1
  xa=np.cos(q1)
  ya=np.sin(q1)
  xb=np.cos(q1+q2)+np.cos(q1)
  yb=np.sin(q1+q2)+np.sin(q1)
  # return (xa,ya),(xb,yb)
  plt.plot([0,xa],[0,ya],color='red')
  plt.plot([xa,xb],[ya,yb],color='blue')
  plt.scatter([0,xa,xb],[0,ya,yb],color='black')
  plt.text(0,0,'Origin',fontsize=10)
  plt.text(xa,ya,'Elbow Joint',fontsize=10)
  plt.text(xb,yb,'End Effector',fontsize=10)
  plt.show()
```

```python
computation_plot(np.pi/3,0)
```

```python
computation_plot(np.pi/3,np.pi/4)
```

```python
computation_plot(np.pi/3,179*np.pi/180) #q2=179 degree
```

```python
computation_plot(np.pi/3,np.pi) ##q2=180 degree
```
